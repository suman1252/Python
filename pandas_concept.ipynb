```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
        'A': [1, 2, 2, 3, 3, 3, 4, 4, 4, 4],
        'B': [2, 3, 4, 4, 5, 5, 5, 6, 6, 6]
    }
df = pd.DataFrame(data)
# def plot_histogram(df, 'A', bins=5)


def plot_histogram(df, column_name, bins=10):
    """
    Plots a histogram for a given numeric column in a DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    column_name (str): The name of the numeric column to plot.
    bins (int): The number of bins to use in the histogram. Default is 10.
    """
    if column_name not in df.columns:
        print(f"Column '{column_name}' not found in DataFrame.")
        return
    
    if not pd.api.types.is_numeric_dtype(df[column_name]):
        print(f"Column '{column_name}' is not numeric.")
        return

    # Plotting the histogram
    # plot_histogram(df, 'A', bins=5)
    plt.figure(figsize=(10, 6))
    plt.hist(df[column_name], bins=bins, edgecolor='black')
    plt.title(f'Histogram of {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    plot_histogram(df, 'A', bins=5)
```

```python
import pandas as pd

data = {
    'date': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'value': range(100)
}
df = pd.DataFrame(data)

def resample_time_series(df, date_column, value_column, new_frequency, agg_func='mean'):
    """
    Resamples time series data in a DataFrame to a new frequency.

    Parameters:
    df (pd.DataFrame): DataFrame containing the time series data.
    date_column (str): Name of the column containing date/time values.
    value_column (str): Name of the column containing the values to be resampled.
    new_frequency (str): New frequency for resampling (e.g., 'D' for daily, 'M' for monthly).
    agg_func (str or callable): Aggregation function to use for resampling (e.g., 'mean', 'sum').

    Returns:
    pd.DataFrame: DataFrame with resampled time series data.
    """
    # Ensure the date column is in datetime format
    df[date_column] = pd.to_datetime(df[date_column])

    # Set the date column as the index
    df.set_index(date_column, inplace=True)

    # Resample the data
    resampled_df = df.resample(new_frequency).agg({value_column: agg_func})

    # Reset the index to turn the date back into a column
    resampled_df.reset_index(inplace=True)

    return resampled_df

# Example usage:
# Sample data
# data = {
#     'date': pd.date_range(start='2023-01-01', periods=100, freq='D'),
#     'value': range(100)
# }
# df = pd.DataFrame(data)

# Resample daily data to monthly frequency using the mean aggregation function
resampled_df = resample_time_series(df, date_column='date', value_column='value', new_frequency='M', agg_func='mean')

print(resampled_df)
```

```python
import pandas as pd

def impute_missing_values(df, strategy="mean", columns=None):
  """
  Imputes missing values in a DataFrame using different strategies.

  Args:
      df (pd.DataFrame): The DataFrame containing missing values.
      strategy (str, optional): The imputation strategy to use. Can be "mean", "median", "mode", or "ffill" (forward fill). Defaults to "mean".
      columns (list, optional): A list of column names to impute. If None, imputes all columns. Defaults to None.

  Returns:
      pd.DataFrame: The DataFrame with imputed missing values.
  """
  
  if columns is None:
    columns = df.columns

  imputed_df = df.copy()

  for col in columns:
    if strategy == "mean":
      imputed_df.loc[imputed_df[col].isnull(), col] = df[col].mean()
    elif strategy == "median":
      imputed_df.loc[imputed_df[col].isnull(), col] = df[col].median()
    elif strategy == "mode":
      imputed_df.loc[imputed_df[col].isnull(), col] = df[col].mode()[0]  # Assuming unique mode
    elif strategy == "ffill":
      imputed_df[col] = imputed_df[col].fillna(method="ffill")  # Forward fill
    else:
      raise ValueError(f"Invalid imputation strategy: {strategy}")

  return imputed_df
```

```python
import pandas as pd

def reshape_and_average_scores(df):
  """
  Reshapes a DataFrame with subjects and student scores to have subjects as columns and students as rows with average scores.

  Args:
      df (pd.DataFrame): The DataFrame containing student scores.

  Returns:
      pd.DataFrame: The reshaped DataFrame with average scores per subject.
  """

  # Assuming 'Subject' and 'Student' are column names for subject and student information
  # and other columns contain student scores (e.g., 'Score1', 'Score2', ...)
  score_columns = [col for col in df.columns if col not in ['Subject', 'Student']]

  # Reshape with pivot_table (alternative: set_index + unstack)
  average_scores = df.pivot_table(index='Student', columns='Subject', values=score_columns, aggfunc='mean')

  return average_scores

# Example Usage (assuming you have your DataFrame loaded as 'df')
average_scores_df = reshape_and_average_scores(df.copy())
print(average_scores_df)
```
